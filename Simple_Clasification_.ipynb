```
#converting the Dependent column in to Category type , so that we can classify the label from all our Independent columns
train_data[65]=train_data[65].astype('category')
```

```
train_data[64].describe()
```

```
# using label encoder to column number 64 as it is order wise categry values
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
train_data[64] = le.fit_transform(train_data[64])
```

```
train_data[64].describe()
```

```
train_data[65].describe()
```

```
#taking all independent columns in to X variable
X=train_data.loc[:, train_data.columns != 65]
```

```
#label column in to y variable

y=train_data[65]
```

```
#splitting our data in to train and test split to check the accuracy
from sklearn.model_selection  import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```
#used logistic regression to predict the label column
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn import metrics
logsk = LogisticRegression()
logsk.fit(X_train,y_train)
y_pred_class = logsk.predict(X_test)
print(classification_report(y_test, y_pred_class))
```
